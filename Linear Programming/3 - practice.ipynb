```python
import matplotlib.pyplot 
from matplotlib.pyplot import *  
import numpy  
from numpy import arange 
figure()  
x= arange(0.0,10000.0,0.1)  
y= arange(0.0,10000.0,0.1)
y1= 9000.0 - 3.0*x  
y2= 3000.0 - .75*x  
xlim(0,10000)  
ylim(0,10000)  
xlabel('x-axis')  
ylabel('y-axis')  
plot(x,y1, c='r', label='y=9000-3x')  
plot(x,y2, c='b', label='y=3000-1.33x')  
legend(['y1=9000-3x','y2=3000-1.33x'],loc='best')  
fill_between(x,y2,where=(y2<=y1),  color= 'b') 
fill_between(x,y1,where=(y1<=y2), color= 'b') 
title ('Shaded Area Shows the Feasible Region')  
show()  
```

```python
import numpy 
from numpy import * 
from numpy.linalg import * 
 
x= [6,6,4,0]  
y= [6,3,2,3]  
obj= matrix([1.0,10.0])  
obj= transpose(obj)  
corners= matrix([x,y])  
corners= transpose(corners)  
result= dot(corners,obj)  
print ('Value of Objective Function at Each Corner Point:\n', result) 
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import *  
import numpy  
from numpy import arange 
figure()  
x= arange(0.0,100.0,0.1)  
#y1= arange(0.0,20.0,0.1)
#y= arange(0.0,40.0,0.1)
y1= 45.0 - 1.0*x  
y2= 30.0 - 1*x 
y3= 0.66*x
xlim(10,60)  
ylim(0,60)  
xlabel('x-axis')  
ylabel('y-axis')  
#plot(x,x1, c='r', label='x=10')
plot(x,y1, c='r', label='y1=45.0 - 1.0*x')  
plot(x,y2, c='b', label='y2=30.0 - 1*x') 
plot(x,y3, c='g', label='y2=30.0 - 1*x')

legend(['y1=45-x','y2=30-x'],loc='best')  
fill_between(x,y2,y1,where=((y2>y3)), color= 'b')  
fill_between(x,y3,y1,where=(y3>=y2)&(y3<y1), color= 'b') 
title ('Shaded Area Shows the Feasible Region')  
show()  
```
